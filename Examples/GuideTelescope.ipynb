### Plotting all the telescopes together

```python
import matplotlib.pyplot as plt
def plot_telescopes(telescopes):
    """
    Plots the acceptance of multiple telescope objects in subplots.
    
    Parameters:
    telescopes (list): List of telescope objects.
    """
    # Create a figure with a grid of subplots
    n = len(telescopes)
    fig, axs = plt.subplots(n, 1, figsize=(10, 5*n))

    # If there's only one telescope, axs is not a list
    if n == 1:
        axs = [axs]

    # Plot each telescope's acceptance in a different subplot
    for i, telescope in enumerate(telescopes):
        ax = axs[i]
        # Call the acceptance method of the telescope object
        # This should plot the acceptance on the given axis
        telescope.acceptanceplot()
        ax.set_title(telescope.name)

    plt.tight_layout()
    plt.show()
```

```python
def plot_telescopes(telescopes):
    """
    Plots the acceptance of multiple telescope objects in subplots.
    
    Parameters:
    telescopes (list): List of telescope objects.
    """
    # Create a figure with a grid of subplots
    n = len(telescopes)
    nrows = 2
    ncols = 3
    fig, axs = plt.subplots(nrows, ncols, figsize=(10*ncols, 5*nrows))

    # If there's only one telescope, axs is not a list
    if n == 1:
        axs = [axs]

    # Plot each telescope's acceptance in a different subplot
    for i, telescope in enumerate(telescopes):
        ax = axs[i//ncols, i%ncols]
        # Call the acceptance method of the telescope object
        # This should plot the acceptance on the given axis
        telescope.acceptanceplot()
        ax.set_title(telescope.name)

    plt.tight_layout()
    plt.show()
```

```python
telescopes = [my_telescopeLesparre, my_telescopeLesparre2, my_telescopePresi, my_telescopeCarbone, my_telescopeGilbert, my_telescopeUchida]
plot_telescopes(telescopes)
```
